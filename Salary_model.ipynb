```python
import pandas as pd

df = pd.read_csv('hiring (1).csv')
df.head()
```

```python
#Convert words to numbers
exp_map = {
    'zero':0, 'one':1, 'two':2, 'three':3,
    'four':4, 'five':5, 'six':6, 'seven':7,
    'eight':8, 'nine':9, 'ten':10
}

df['experience'] = df['experience'].map(exp_map)
```

```python
#Fill missing values
df['experience'].fillna(0, inplace=True)
df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].median(),
    inplace=True
)
```

```python
df
```

```python
X = df[['experience',
        'test_score(out of 10)',
        'interview_score(out of 10)']]

y = df['salary($)']
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X, y)
```

```python
predictions = lm.predict([[2, 9, 6],[12, 10, 10]])

import numpy as np
np.round(predictions, 2)
```
